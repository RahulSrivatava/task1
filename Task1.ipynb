<p style="color:yellow"><b>Importing Require Libraries</b></p>

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

<p style="color:yellow"><b>Importing Data Set</b></p>

```python
data_set = pd.read_csv( "http://bit.ly/w-data")
print("Data imported successfully")

data_set.head(5)
```

<p style="color:yellow"><b>Check for Null Values</b></p>

```python
data_set.isnull().sum()
```

<p style="color:yellow"><b>Shape of Data Set</b></p>

```python
data_set.shape
```

<p style="color:yellow"><b>Visualising Data Set to get the insights of Data Set </b></p>

```python
data_set.plot(x='Hours', y='Scores', style='x',color='Red')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

<p style="color:yellow"><b>Dividing Data Set into Dependent & Independent Variable </b></p>

```python
x =data_set.iloc[:, :-1].values  
y = data_set.iloc[:, 1].values  
```

<p style="color:yellow"><b>Now i have splited the data set by Scikit-Learn's built-in Function tain_test_split </b></p>

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                            test_size=0.3, random_state=0) 
```

<p style="color:yellow"><b>Traing of Algorithm </b></p>

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(x_train, y_train) 

print("Training complete.")
```

 <p style="color:yellow"><b>Ploting Regression line for the Test Data </b></p>

```python
line = regressor.coef_*x+regressor.intercept_
plt.scatter(x, y,color="purple")
plt.plot(x, line,color="Red");
plt.show()
```

 <p style="color:yellow"><b>Now It's time to Make some Pridiction </b></p>

```python
print(x_test) 
y_pred = regressor.predict(x_test) 
```

```python
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
comparison 
```

<h2><b>Problem statement</b></h2>
 <p style="color:yellow">What will be predicted score if a student studies for 9.25 hrs/ day?</p>

```python
hours = [[9.25]]
problem = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(problem[0]))
```

<h2><p style="color:yellow"><b>Conclusion</b></p></h2>
<h3>So, Now I conclude that if a student studies for 9.25hrs/day he will be getting score of 92.91%</h3>
